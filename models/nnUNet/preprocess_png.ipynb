```python
from PIL import Image
from glob import glob
import numpy as np
import cv2
```

```python
data_home = 'C:/Users/qwe14/0.code/sarcopenia/data_train/'
annot = "train_annot2/annot_train.txt"
annot_file = f"{data_home}{annot}"

with open(annot_file, "r", encoding='utf-8') as f:
    lines = f.readlines()

f_names_train = []
f_names_test = []
for idx, line in enumerate(lines):
    f_n = line.strip().split(" ")[0]
    if idx % 3 == 0:
        if idx % 30 == 0:
            f_names_test.append(f_n[1:])
        else:
            f_names_train.append(f_n[1:])


print(len(lines))
print(len(f_names_train))
print(len(f_names_test))
```

```python
data_home = 'C:/Users/qwe14/0.code/sarcopenia/data_train/'
image_home = f"{data_home}images_2/"

new_image_home = "C:/Users/qwe14/0.code/nnUNet/nnUNet_raw/Dataset121_MR/imagesTr/"
for img_file in f_names_train:
    img_path = f"{image_home}{img_file}"
    p_image = Image.open(img_path)
    img_file = img_file.replace("jpg", "png")
    new_annot_img = f"{new_image_home}{img_file}"
    p_image.save(new_annot_img)
```

```python
data_home = 'C:/Users/qwe14/0.code/sarcopenia/data_train/'
image_home = f"{data_home}images_2/"

new_image_home = "C:/Users/qwe14/0.code/nnUNet/nnUNet_raw/Dataset121_MR/imagesTs/"
for img_file in f_names_test:
    img_path = f"{image_home}{img_file}"
    p_image = Image.open(img_path)
    img_file = img_file.replace("jpg", "png")
    new_annot_img = f"{new_image_home}{img_file}"
    p_image.save(new_annot_img)
```

```python
data_home = 'C:/Users/qwe14/0.code/sarcopenia/data_train/'
annot_home = f"{data_home}annotations_2/"

new_label_home = "C:/Users/qwe14/0.code/nnUNet/nnUNet_raw/Dataset121_MR/labelsTr/"
for img_file in f_names_train:
    init_SVM = ["S", "V", "M"]
    # 피하지방(S): 1
    # 내장지방(V): 2
    # 근육(M): 3
    mask_final = np.zeros((512, 512))
    for idx, initial in enumerate(init_SVM):
        mask_path = f"{annot_home}{initial}{img_file}"
        p_image = Image.open(mask_path)
        np_image = np.array(p_image)
        np_image = np.mean(np_image, axis = 2, dtype=np.int32)
        np_image[np_image< 128] = 0
        np_image[np_image >= 128] = 255
        mask_final[np_image == 255] = idx + 1
    img_file = img_file.replace("jpg", "png")
    new_annot_label = f"{new_label_home}{img_file}"
    cv2.imwrite(new_annot_label, mask_final)
```
